```python
def df_explore(df):
    print('Shape:', df.shape, end='\n\n')
    
    for col in list(df):
        print('Column:', col)
        print('Count:', df[col].count())
        print('%NaN:', round((df.shape[0] - df[col].count())/df.shape[0], 4))
        print('Unique Entries:',  len(df[col].unique()))
        print('Value Counts:')
        print(df.groupby(col).size(), end='\n\n')

df_explore(df)
```
